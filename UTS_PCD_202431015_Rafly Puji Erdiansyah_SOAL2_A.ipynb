## Memanggil Library

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Membaca Gambar dan Konversi ke RGB

```python
#Membaca gambar dan merubah ke RGB
image = cv2.imread('Gambar2 backlight foto.jpeg')
img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(6,8))
plt.imshow(img_rgb)
plt.title('Gambar Asli')
plt.show()
```

## Merubah gambar ke GrayScale

```python
#Mengubah gambar ke format GrayScale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(6,8))
plt.imshow(gray, cmap='gray')
plt.title('Gambar Gray')
plt.show()
```

## Menambah kecerahan Gambar GrayScale

```python
#Menambah kecerahan gambar GrayScale
kecerahan = 60
gray_bright = cv2.add(gray, kecerahan)

plt.figure(figsize=(6,8))
plt.imshow(gray_bright, cmap='gray')
plt.title('Gambar Gray yang dipercerah')
plt.show()
```

## Menambahkan Cahaya pada gambar GrayScale

```python
#Menambahkan cahaya pada gambar GrayScale
alpha = 2.0 
gray_contrast = cv2.convertScaleAbs(gray, alpha=2.0, beta=0)

plt.figure(figsize=(6,8))
plt.imshow(gray_contrast, cmap='gray')
plt.title('Gambar Gray yang diperkontras')
plt.show()
```

## Menambahkan Contras dan Cahaya pada gambar GrayScale

```python
#Menambahkan Contras dan Cahaya pada gambar GrayScale
beta = 60
alpha = 2.0
gray_BC = cv2.convertScaleAbs(gray, alpha=2.0, beta=60)

plt.figure(figsize=(6,8))
plt.imshow(gray_BC, cmap='gray')
plt.title('Gambar Gray yang dipecerah dan diperkontras')
plt.show()
```
